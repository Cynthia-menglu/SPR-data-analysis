```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IgEFc_to_FceRIa binding data.csv')
df
```

```python
#Drop null value columns to avoid error
df.dropna(inplace = True)
df
```

```python
#rename the columns
df.rename(columns={'Column1': 'T', 'Column2': '2e-3nM', 'Column3': '2e-2nM', 'Column4': '2e-1nM', 'Column5': '2nM', 'Column6': '2e1nM', 'Column7': '2e2nM', 'Column8': '2e3nM'}, inplace=True)
df
```

```python
#convert dtypes of data into float
for r in range(df.shape[1]):
    df[df.columns[r]] = pd.to_numeric(df[df.columns[r]], errors = 'coerce')
df.dtypes
```

```python
#delete baseline values
for i in df.index:
    if df.loc[i, 'T'] < 0:
        df = df.drop(index = i)
    else:
        break
df
```

```python
#Set the time t for association
t = input('Time for association: ')
```

```python
t = float(t)
```

```python
for i in range(df.shape[0]):
    if df.iloc[i, 0] == t:
        tp = df[df[df.columns[0]] == df.iloc[i, 0]].index.tolist()[0]
        print(tp)
```

```python
#Build two new dataframes for association and dissociation data
df_association = df.loc[600:tp, 'T':'2e3nM']
df_dissociation = df.loc[tp:12000, 'T':'2e3nM']
df_association.to_csv('Association binding data.csv')
df_dissociation.to_csv('Dissociation data.csv')
```
